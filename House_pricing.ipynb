```python
def isfloat(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
df2[~df2['total_sqft'].apply(isfloat)].head(10)
```

```python
def convert_sqft_to_num(x):
  tokens=x.split('-')
  if len(tokens)==2:
    return(float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df3=df2.copy()
df3['total_sqft']=df3['total_sqft'].apply(convert_sqft_to_num)
df3=df3[df3.total_sqft.notnull()]
df3.head()
```
